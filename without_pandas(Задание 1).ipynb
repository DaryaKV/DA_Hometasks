```python
import csv
import sys
```

```python
csv.field_size_limit(sys.maxsize)
```

```python
vacs = list(csv.DictReader(open('vacancy_nh.csv')))
```

Сколько всего вакансий?

```python
print(len(vacs))
```

За какой период эти вакансии?

```python
dates = []
for i in range(len(vacs)):
  date = vacs[i]['created_at']
  if(date not in dates):
    dates.append(date)
for date in dates:
  print(date)
```

Есть ли вакансии менеджеров с использованием Python, если да сколько их?

```python
i = 0
for vac in vacs:
  if('менеджер' in vac['vactitle'].lower() and 'python' in vac['vactitle'].lower()):
    print(vac['vactitle'])
    i = i + 1
print(str(i) + ' вакансий')
```

Сколько вакансий для аналатика данных?

```python
i = 0
for vac in vacs:
  if('аналитик данных' in vac['vactitle'].lower() or 'data analyst' in vac['vactitle'].lower()):
    i = i + 1
print(str(i) + ' вакансии')
```

Сколько вакансий для аналитика данных с использованием Python?

```python
i = 0
for vac in vacs:
  if(('аналитик данных' in vac['vactitle'].lower() or 'data analyst' in vac['vactitle'].lower()) and 'python' in vac['vactitle'].lower()):
    i = i + 1
print(str(i) + ' вакансия')
```
